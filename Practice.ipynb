```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_filtered["Patient Age at Treatment"]
def age_to_group(age):
    if age == "18 - 34":
        return 0
    elif age == "35-37":
        return 1
    elif age == "38-39":
        return 2
    elif age == "40-42":
        return 3
    elif age == "43-44":
        return 4
    elif age == "45-50":
        return 5
    else:
        return None  # 그 외 값은 None으로 처리
df_filtered['Age Group'] = df_filtered['Patient Age at Treatment'].apply(age_to_group)
# df_filtered["Patient Age at Treatment"] = df_filtered["Patient Age at Treatment"].apply(age_to_group)
```

```python
# df_filtered["Patient Age at Treatment"].astype(int)
df_filtered['Patient Age at Treatment'] = df_filtered['Age Group']
df_filtered.drop(columns="Age Group")
```

```python
df_cleaned = df_filtered[~df_filtered.applymap
(lambda x: str(x).startswith('>=5') or str(x).startswith('> 50')).any(axis=1)]
```

```python
df_cleaned['Total Number of Previous IVF cycles'] = pd.to_numeric(df_cleaned['Total Number of Previous IVF cycles'], errors='coerce')
df_cleaned['Total number of IVF pregnancies'] = pd.to_numeric(df_cleaned['Total number of IVF pregnancies'], errors='coerce')
df_cleaned['Fresh Eggs Collected'] = pd.to_numeric(df_cleaned['Fresh Eggs Collected'], errors='coerce')
df_cleaned['Eggs Mixed With Partner Sperm'] = pd.to_numeric(df_cleaned['Eggs Mixed With Partner Sperm'], errors='coerce')

df_cleaned = df_cleaned.fillna(df_cleaned.mean())  
```

```python
df_cleaned
```

```python
from sklearn.model_selection import train_test_split

y = df_cleaned["Number of Live Births"]
x = df_cleaned.drop(columns=['Number of Live Births'])

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
x_test
```

# XGBOOST 학습

```python
import xgboost as xgb
from xgboost import XGBClassifier
import matplotlib.pyplot as plt
from xgboost import plot_importance
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_breast_cancer

model = XGBClassifier(  
    scale_pos_weight=2, #양성 클래스 가중치
    n_estimators=1000,   # 최대 1000번 학습
    max_depth=10,         # 트리 깊이
    learning_rate=0.1,   # 학습률
    objective='binary:logistic',  # 이진 분류 문제
    eval_metric='logloss',        # 평가 지표로 logloss 사용
    random_state=42)

# 모델 학습 (early stopping 적용)
xgb_model = model.fit(
    x_train, y_train, 
    eval_set=[(x_test,y_test)]   # 검증 데이터셋 (X_val, y_val)
)

# 예측하기
y_pre = xgb_model.predict(x_test)
y_pred_probs = xgb_model.predict_proba(x_test)[:,1]

# 특성 중요도 시각화
fig, ax = plt.subplots(figsize=(10,12))
plot_importance(xgb_model, ax=ax)
plt.show()
```

```python
from sklearn.metrics import f1_score, accuracy_score

f1 = f1_score(y_test, y_pre)
print(f'F1 Score: {f1:.4f}')

accuracy = accuracy_score(y_test, y_pre)
print(f'Accuracy: {accuracy:.4f}')
```
