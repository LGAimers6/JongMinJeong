```python
#시술 당시 나이 처리
def age_encoding(age):
    if age == "만18-34세":
        return 0
    elif age == "만35-37세":
        return 1
    elif age == "만38-39세":
        return 2
    elif age == "만40-42세":
        return 3
    elif age == "만43-44세":
        return 4
    elif age == "만45-50세":
        return 5
    else:
        return None  # 그 외 값은 None으로 처리

train_filtered.loc[:,'시술 당시 나이'] = train_filtered['시술 당시 나이'].apply(age_encoding)
```

```python
train_filtered['시술 당시 나이'].value_counts(dropna=False)
```

```python
train_filtered.loc[:,'특정 시술 유형'] = train_filtered['특정 시술 유형'].fillna("Unknown")
```

```python
train_filtered['특정 시술 유형'].value_counts(dropna=False)
```

```python
#시술 유형 - 특정 시술 유형 원 핫 인코딩 
#AdaBoost 계열은 결정 트리 기반이라 범주형 변수 처리에 강점이 있어 원-핫 인코딩된 변수들도 잘 처리할 수 있다고 기대 
treatment_type = pd.get_dummies(train_filtered['시술 유형'])
specific_type = pd.get_dummies(train_filtered['특정 시술 유형'])

train_filtered = train_filtered.drop(['시술 유형', '특정 시술 유형'], axis=1)
```

```python
df = pd.concat([train_filtered, treatment_type, specific_type], axis=1)
df
```

```python
df['시술 당시 나이'].value_counts(dropna = False)
```

```python
df.loc[:,'시술 당시 나이'] = df['시술 당시 나이'].fillna(1)
```

```python
df['배란 자극 여부'].value_counts(dropna=False)
```

```python
df[['이식된 배아 수']]
df = df.loc[:, ~df.columns.duplicated()]
```

```python
df['이식된 배아 수'].value_counts(dropna=False)
```

```python
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=3)
```

```python
df['이식된 배아 수'] = imputer.fit_transform(df[['이식된 배아 수']])
```

```python
df['이식된 배아 수'].value_counts(dropna=False)
```

```python
def removeHangual(tmp):
    return int(tmp[0])
df.loc[:,['총 시술 횟수']] = df['총 시술 횟수'].apply(removeHangual)
df.loc[:,['클리닉 내 총 시술 횟수']] = df['클리닉 내 총 시술 횟수'].apply(removeHangual)
df.loc[:,['IVF 시술 횟수']] = df['IVF 시술 횟수'].apply(removeHangual)
df.loc[:,['DI 시술 횟수']] = df['DI 시술 횟수'].apply(removeHangual)
df.loc[:,['총 임신 횟수']] = df['총 임신 횟수'].apply(removeHangual)
df.loc[:,['IVF 임신 횟수']] = df['IVF 임신 횟수'].apply(removeHangual)
df.loc[:,['DI 임신 횟수']] = df['DI 임신 횟수'].apply(removeHangual)
df.loc[:,['총 출산 횟수']] = df['총 출산 횟수'].apply(removeHangual)
df.loc[:,['IVF 출산 횟수']] = df['IVF 출산 횟수'].apply(removeHangual)
df.loc[:,['DI 출산 횟수']] = df['DI 출산 횟수'].apply(removeHangual)

```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df.loc[:,'난자 출처'] = encoder.fit_transform(df['난자 출처'])
df.loc[:,'정자 출처'] = encoder.fit_transform(df['정자 출처'])
df.loc[:,'난자 기증자 나이'] = encoder.fit_transform(df['난자 기증자 나이'])
df.loc[:,'정자 기증자 나이'] = encoder.fit_transform(df['정자 기증자 나이'])
```
